```python
import pandas as pd
import seaborn as sns
```

```python
sns.set_style("darkgrid")
titanic_data = sns.load_dataset('titanic')
titanic_data.head(20)
```

```python
# selecting data using brackets[]

print(titanic_data["class"])
type(titanic_data["class"])
```

```python
# printting data with list of columns 
print(type(titanic_data[["class", "age", "sex"]]))
titanic_data[["class", "age", "sex"]]
```

```python
my_df = titanic_data[titanic_data["sex"] == 'male']
my_df
```

```python
# we can add more codition to filter the data

my_df = titanic_data[(titanic_data["sex"] == 'male') & (titanic_data["class"] == "First")]
my_df.head(20)

my_df1 = titanic_data[(titanic_data["alive"] == 'yes') & (titanic_data["class"] == 'First')]
my_df1
```

```python
# use isin() function to specify a range of values

ages = [20, 21, 2]
age_dataset = titanic_data[titanic_data["age"].isin(ages)]
age_dataset.head(10)
```

```python
# indexing and slicing using loc Function

scores = [
{'Subject':'Mathematics', 'Score': 85 , 'Grade': 'B', 'Remarks': 'Good', },
{'Subject':'History', 'Score': 98 , 'Grade': 'A','Remarks': 'Excellent'},
{'Subject':'English', 'Score': 76 , 'Grade': 'C','Remarks': 'Fair'},
{'Subject':'Science', 'Score': 72 , 'Grade': 'C','Remarks': 'Fair'},
{'Subject':'Arts', 'Score': 95 , 'Grade': 'A','Remarks': 'Excellent'},
]

my_df = pd.DataFrame(scores)
my_df.head()
```

```python
print(my_df.loc[2])
type(my_df.loc[2])
```

```python
# we can slice the result my using index number

my_df.loc[2:4]
```

```python
# we can also use column to fliters

my_df.loc[2:4,["Subject", "Grade"]]
```
